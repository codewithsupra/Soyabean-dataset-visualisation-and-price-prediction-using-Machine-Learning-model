```
# Import data analysis tools 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```
df = pd.read_csv("a.csv")
```

```
#fig, ax = plt.subplots()
plt.scatter(df["Area_Under_Soybean_Cultivation"][:], df["Production_of_Soybean"][:])
plt.title("Area vs. Production")
plt.xlabel("Area Under Soybean Cultivation")
plt.ylabel("Production of Soybean")
```

```
df.Production_of_Soybean.plot.hist()
```

### Linear Regression
Predict the production of Soyabean using Linear Regression for different Area


```
df_temp = df.drop("Year", axis=1)
x = df_temp.iloc[:,:-1]
y = df_temp.iloc[:,1]
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x, y)
```

```
x_test = [[11],[12],[13],[14],[15]]
y_pred = reg.predict(x_test)
y_pred
```

```
plt.scatter(x, y, color = 'green')
plt.plot(x, reg.predict(x), color = 'blue')
plt.title("Area vs Production")
plt.xlabel('Area_Under_Soybean_Cultivation')
plt.ylabel('Production_of_Soybean')
plt.show()
```

# B

```
df = pd.read_csv("b.csv")
df.head()
```

```
plt.scatter(df["Quantity"][:], df["Value"][:])
plt.title("Quantity vs. Value")
plt.xlabel("Quantity")
plt.ylabel("Value")
```

```
fig1, ax1 = plt.subplots()
ax1.pie(df["Quantity"], autopct='%1.1f%%',
        shadow=False)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Distribution of the Quantity")
plt.show()
```

### Regresion

```
df_temp = df.drop("Financial_Year", axis=1)
x = df_temp.iloc[:,:-1]
y = df_temp.iloc[:,1]
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x, y)
x_test = [[5],[6],[7],[7.5],[8]]
y_pred = reg.predict(x_test)
y_pred
```

```
plt.scatter(x, y, color = 'green')
plt.plot(x, reg.predict(x), color = 'blue')
plt.title("Quantity vs. Value")
plt.xlabel("Quantity")
plt.ylabel("Value")
plt.show()
```
